```python
#Importing the packages

import pandas as pd
import numpy as np
import missingno as mn
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Definignt he column headings 
cols = ["Age","Workclass","Final_Weight","Education","Education_Num","Marital_Status","Occupation","Relationship","Race","Sex","Capital_Gain","Capital_Loss","Hours_Per_Week","Native_Country","Income"]

# Reading the data to dataframe
df = pd.read_csv('adult.data',names=cols,skipinitialspace=True)
df
```

```python
#printing first five rows using head() method
df.head()
```

```python
#Printing last five rows using tail() method
df.tail()
```

```python
#Getting the DataFrame Column Charecteristics using info()
df.info()
```

# HANDLING THE MISSING VALUES

```python
#Checking for NUll Values or Missing Values
df.isnull().sum()
```

```python
#It is Observed in the data set that some missing values are replaced by '?'
df.isin(['?']).sum()
```

```python
#Replacing the '?' with 'Nan'
df['Workclass'] = df['Workclass'].replace('?', np.nan)
df['Occupation'] = df['Occupation'].replace('?', np.nan)
df['Native_Country'] = df['Native_Country'].replace('?', np.nan)
```

```python
#Checking for NUll Values or Missing Values
df.isnull().sum()
```

```python
#Dropping the Null Values
df.dropna(inplace=True)
```

```python
#Checking for Duplicate Records
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

## BIVARIATE ANALYSIS

```python
#Relationship between Hours_Per_Week and Income using box Plot

sns.boxplot(x = 'Income', y = 'Hours_Per_Week', data=df[cols])
```

It is Observed that people who work for 40 hrs to 50 Hrs a week are likely to earn above 50K but on the other hand there are some outliers who work more than 50 hrs a week but still income is less than 50k. 

We conclude here that Hours_Per_week and Income are not linearly releated.

```python
df[(df["Income"]==0) & (df["Hours_Per_Week"]>50)].groupby(["Occupation"]).agg({"Hours_Per_Week":"mean"})
```

Group of people working more then 50hrs and earning less income are from Priv-house-serv occupation

```python
df[(df["Income"]==1) & (df["Hours_Per_Week"]<50)].groupby(["Occupation"]).agg({"Hours_Per_Week":"mean"})
```

People who earn more than 50K are mostly working 40 hrs per week

```python
tot_lowpaying = df[df["Income"]==0].shape[0]
num_hardworking = df[(df["Income"]==0) & (df["Hours_Per_Week"]>=50)].shape[0]
hardworking_ppl = (num_hardworking/tot_lowpaying) * 100
```

```python
print(f"there are a total of {tot_lowpaying} people who earn less than 50K, \n in which number of people who work more than 50hrs are {num_hardworking} and their proportion is {hardworking_ppl}")
```

```python
tot_highpaying = df[df["Income"]==1].shape[0]
num_smart = df[(df["Income"]==1) & (df["Hours_Per_Week"]<50)].shape[0]
smartworking_ppl = (num_smart/tot_highpaying) * 100
```

```python
print(f"there are a total of {tot_highpaying} people who earn more than 50K, \n in which number of people who work less than 50hrs are {num_smart} and their proportion is {smartworking_ppl}")
```

The difference in Capital_Gain and Capital_Loss gives us the profit/loss. Let's create a new column and replace them with
these two columns

```python
df["Profit/Loss"] = df["Capital_Gain"] - df["Capital_Loss"]
df.drop(["Capital_Gain","Capital_Loss"],inplace=True,axis=1)
```

```python
sns.pairplot(df)
```

since there are many categories with categorical data, using label encoder just to see how they are linearly related to Income
and not using one hot encoding since that will result in too many columns 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = df.apply(LabelEncoder().fit_transform)
```

```python
df.head()
```

```python
corr = df.corr()
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(corr,annot=True)
```

It turns out that none of the features have high correlation with income. Only Age, Education_Num, Relationship, Sex,
Hours and profit/Loss have low correlation. Interestingly Workclass and Occupation ain't showing any linear relationship
with Income. Race, Final_weight and Native_Country are poor predictors of Income so they can be dropped from further analysis.
